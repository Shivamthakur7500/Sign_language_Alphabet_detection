```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
with open("/content/sign_templates.json", "r") as f:
      data = json.load(f)
```

```python
keys=[]
value=[]
for key, values in data.items():
   keys.append(key)
   value.append(values)
for i in range(len(keys)):
  print(f"keys= {keys[i]} value= {value[i]}")
```

```python
size=[]
for i in range(len(value)):
    size.append(len(value[i]))
    maxsize=max(size)
maxsize
```

```python

for i in range(len(value)):
    flat = np.array(value[i])
    value[i] = flat.flatten()
    print(value[i])
```

```python
for i in range(len(value)):
  print(value[i][:4])
```

```python
data={
    "keys":keys,
    "values":value
}
df=pd.DataFrame(data)
df
```
